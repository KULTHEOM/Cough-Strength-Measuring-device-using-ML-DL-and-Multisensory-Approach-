```python
#pseudocough
#mild
#unknown
#severe
def severity_report_mild(row):
    if pd.notna(row['severity_1']):
        if row['severity_1']  == 'mild':
            return 2
        else:
            return 0
    if pd.notna(row['severity_2']):
        if row['severity_2']=='mild':
            return 2
        else:
            return 0
    if pd.notna(row['severity_3']):
        if row['severity_3']=='mild':
            return 2
        else:
            return 0
    if pd.notna(row['severity_4']):
        if row['severity_4']=='mild':
            return 2
        else:
            return 0
        
def severity_report_severe(row):
    if pd.notna(row['severity_1']):
        if row['severity_1']  == 'severe':
            return 3
        else:
            return 0
    if pd.notna(row['severity_2']):
        if row['severity_2']=='severe':
            return 3
        else:
            return 0
    if pd.notna(row['severity_3']):
        if row['severity_3']=='severe':
            return 3
        else:
            return 0
    if pd.notna(row['severity_4']):
        if row['severity_4']=='severe':
            return 3
        else:
            return 0
    
def severity_report_psedocough(row):
        if pd.notna(row['severity_1']):
            if row['severity_1']  == 'pseudocough':
                return 1
            else:
                return 0
        if pd.notna(row['severity_2']):
            if row['severity_2']=='pseudocough':
                return 1
            else:
                return 0
        if pd.notna(row['severity_3']):
            if row['severity_3']=='pseudocough':
                return 1
            else:
                return 0
        if pd.notna(row['severity_4']):
            if row['severity_4']=='pseudocough':
                return 1
            else:
                return 0
    
```

```python
read['mild'] = read.apply(lambda row: severity_report_mild(row), axis=1)
read['severe'] = read.apply(lambda row: severity_report_severe(row), axis=1)
read['pseudocough'] = read.apply(lambda row: severity_report_psedocough(row), axis=1)
```
